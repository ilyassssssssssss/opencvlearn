```python

lake = cv2.imread('images/New_Zealand_Lake.jpg')
plt.imshow(lake)
```

```python
#split the image into its channels
b, g, r = cv2.split(lake)

plt.figure(figsize=[20, 5]) #plt figure size

plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel") 
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel") 
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel")

merged = cv2.merge([b, g, r])
merged = merged[:,:,::-1] #convert to rgb
plt.subplot(144);plt.imshow(merged);plt.title("Merged Image") 
```

```python
#modifying the individual channels
b,g,r = cv2.split(lake)
r = r + 30

plt.figure(figsize=[20, 5]) #plt figure size

plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel") 
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel") 
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel")

merged = cv2.merge([b, g, r])
merged = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)
plt.subplot(144);plt.imshow(merged);plt.title("Merged Image")
```

```python
edited = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)
edited = cv2.imwrite('images/New_Zealand_Lake_edited.jpg', merged)
```
