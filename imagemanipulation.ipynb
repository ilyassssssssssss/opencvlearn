```python
#images using opencv

import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
from IPython.display import Image, display
```

```python
cb = cv2.imread('images/checkerboard_18x18.png')
cb = cv2.cvtColor(cb, cv2.COLOR_BGR2RGB)
plt.imshow(cb, cmap='gray')
```

```python
cb = cv2.cvtColor(cb, cv2.COLOR_BGR2RGB)
cb[2, 3] = 255
cb[2, 2] = 255
cb[2, 1] = 255
plt.imshow(cb, cmap='gray')
```

```python
boat = cv2.imread('images/New_Zealand_Boat.jpg')
boat = cv2.cvtColor(boat, cv2.COLOR_BGR2RGB)
plt.imshow(boat)
```

```python
crop = boat[200:400, 400:600]
plt.imshow(crop)
```

```python
crop_resize = cv2.resize(crop, (100, 100))
#or crop_resize = cv2.resize(crop, (0, 0), fx=5, fy=5)
plt.imshow(crop_resize)
```

```python
boat_horizontal_flip = cv2.flip(boat, 1)
boat_vertical_flip = cv2.flip(boat, 0)
boat_both_flip = cv2.flip(boat, -1)

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1); plt.imshow(boat); plt.title('Original')
plt.subplot(1, 4, 2); plt.imshow(boat_horizontal_flip); plt.title('Horizontal flip')
plt.subplot(1, 4, 3); plt.imshow(boat_vertical_flip); plt.title('Vertical flip')
plt.subplot(1, 4, 4); plt.imshow(boat_both_flip); plt.title('Both flip')
```
