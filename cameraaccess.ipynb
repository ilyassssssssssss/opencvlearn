```python
import cv2
import matplotlib.pyplot as plt
import time

cap = cv2.VideoCapture(0)
time.sleep(1)  # Allow time for the camera to initialize
ret, frame = cap.read()
#print(ret)
#print(frame)

if ret:
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
else:
    print("Failed to capture frame")

cap.release()
plt.show()
```
