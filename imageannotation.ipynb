```python
#images using opencv

import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
from IPython.display import Image, display
```

```python
boat = cv2.imread('images/New_Zealand_Boat.jpg')
boat = cv2.cvtColor(boat, cv2.COLOR_BGR2RGB)
plt.imshow(boat)
```

```python
cv2.line(boat, (0, 0), (boat.shape[1], boat.shape[0]), (255, 0, 0), 5)
plt.imshow(boat)
```

```python
cv2.circle(boat, (boat.shape[1]//2, boat.shape[0]//2), 100, (0, 255, 0), 5)
plt.imshow(boat)
```

```python
cv2.rectangle(boat, (boat.shape[1]//4, boat.shape[0]//4), (3*boat.shape[1]//4, 3*boat.shape[0]//4), (0, 0, 255), 5)
plt.imshow(boat)
```

```python
txt = 'Boat'
cv2.putText(boat, txt, (boat.shape[1]//2, boat.shape[0]//2), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 5)
plt.imshow(boat)
```
